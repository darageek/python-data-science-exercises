# Break and Continue
break: used to exit all coming iterations in loop  
continue: it just skips the current iteration and moves to the nest iteration  
- whenever we wants to terminates the loop iteration that's where break and continue are used

```python
for i in range(1,10):
    if (i==5):
        break
    print(i)
```

```python
# 5 is skipped
for i in range(1,10):
    if (i==5):        continue
    print(i)
```

```python
print(user_info)
```

```python
type(user_info)
```

```python
#accessing elements in dictionary
print(user_info['name'])
print(user_info['nationality'])
```

use dic.get() to avoid error

```python
# no error
user_info.get('hobby')
```

```python
print(user_info)
```

```python
# display all values
user_info.values()
```
